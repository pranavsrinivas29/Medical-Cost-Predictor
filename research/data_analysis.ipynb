```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/insurance.csv")
```

```python
df.head(10)
```

```python
sns.histplot(data=df, x="age", bins=10, kde=True)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

### More people in bmi range of 27 - 35

```python
sns.histplot(data=df, x="bmi", bins=10, kde=True)
plt.xlabel("BMI")
plt.ylabel("Count")
plt.title("BMI Distribution")
plt.show()
```

### Balanced ratio of Gender

```python
sns.countplot(data=df, x="sex")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Count")
plt.show()
```

### Not much skewness in region

```python
region_counts = df["region"].value_counts()

plt.pie(
    region_counts,
    labels=region_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Gender Distribution")
plt.axis("equal")  # makes it a circle
plt.show()
```

### Most people have no kids or 1 kid

```python
children_counts = df["children"].value_counts()

plt.pie(
    children_counts,
    labels=children_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Children Distribution")
plt.axis("equal")  # makes it a circle
plt.show()
```
